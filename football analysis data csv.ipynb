```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data into a pandas DataFrame
df = pd.read_csv(r'C:\Users\shah8\hello\soccer game analysis\s21-22.csv')

x1 = df['Time']
y1 = df['HF']

x2 = df['Time']
y2 = df['AF']

# First plot
plt.scatter(x1, y1)

plt.title('Distribution of  Team Fouls in 21/22 PL')
plt.xlabel('Time')
plt.ylabel('Team Fouls')

# Second plot
plt.scatter(x2, y2)

plt.title('Distribution of Away Team Fouls in 21/22 PL')
plt.xlabel('Time')
plt.ylabel('Team Fouls')

# Display the plots
plt.show()
```
